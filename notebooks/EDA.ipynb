# Converting my categorical columns into numerical columns

```python
categorical_columns
```

```python
df["satisfaction"].unique()
```

```python
gender_map = {"Male":1, "Female":2}
customer_type_map = {"disloyal Customer" : 1, "Loyal Customer" : 2}
type_of_travel_map = {"Personal Travel" : 1, "Business travel":2}
class_map = {"Eco":1 ,"Eco Plus" : 2, "Business":3}
satisfaction_map = {"neutral or dissatisfied":1, "satisfied":2}
```

#### Applying the entire encoded values in my dataframe

```python
df["Gender"] = df["Gender"].replace(gender_map)
df["Customer Type"] = df["Customer Type"].replace(customer_type_map)
df["Type of Travel"] = df["Type of Travel"].replace(type_of_travel_map)
df["Class"] = df["Class"].replace(class_map)
df["satisfaction"] = df["satisfaction"].replace(satisfaction_map)
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool_))
f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(150, 1, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap="bwr", vmax=None, center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .9})
```
